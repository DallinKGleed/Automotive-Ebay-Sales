```python
# importing required libraries
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
```

```python
# path to clean data

clean_data_path = "/Users/dallingleed/Downloads/Capstone Project/cleaned_autos_2.csv"

# reading csv into raw dataframe
df = pd.read_csv(clean_data_path, encoding="latin-1")

df.head()
```

## No of Vehicles by Brand Available on ebay for sale

```python
# Count plot to show the number of vehicles belonging to each brand
sns.set_style("whitegrid")
brands = sns.factorplot(y="Brand", data=df, kind="count", palette="Reds_r", size=7, aspect=1.5, order=df['Brand'].value_counts().index)
brands.ax.set_title("Count of vehicles by Brand",fontdict={'size':18})
brands.ax.xaxis.set_label_text("Count of Vehicles",fontdict= {'size':16})
brands.ax.yaxis.set_label_text("Brand of Vehicles",fontdict= {'size':16})
plt.show()
```

```python
# saving the plot

brands.savefig("/Users/dallingleed/Downloads/Capstone Project/Analysis2Plots/brand-vehicleCount.png")
```

## Average price for vehicles based on the type of vehicle as well as on the type of gearbox

```python
fig, ax = plt.subplots(figsize=(8,5))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
sns.barplot(x="Vehicle Type", y="Price",hue="Gearbox", palette=colors, data=df)
ax.set_title("Average price of vehicles by vehicle type and gearbox type",fontdict= {'size':12})
ax.xaxis.set_label_text("Type Of Vehicle",fontdict= {'size':12})
ax.yaxis.set_label_text("Average Price",fontdict= {'size':12})
plt.show()
```

```python
# saving the plot

fig.savefig("/Users/dallingleed/Downloads/Capstone Project/Analysis2Plots/vehicletype-gearbox-price.png")
```
